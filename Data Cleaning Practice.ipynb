### Duplicates Handling

```python
%%time

# Check for duplicates
duplicated_values = df.duplicated().sum()
print(f"This dataset contains {duplicated_values} duplicated values.")

print()
```

```python
%%time

# Remove duplicates
df = df.drop_duplicates()
df
```

### Remove Unnecessary Columns

```python
# Remove columns that won't be used in analysis
df = df.drop(columns = "Not_Useful_Column")
df
```

### Standardize Data

```python
%%time

# Clean up any special characters, digits, and leading/trailing spaces for Last_Name. 
df["Last_Name"] = (
    df["Last_Name"]
    .str.strip()
    .str.replace(r"[^a-zA-Z\s]", "", regex = True)
    .str.replace(r"\d+", "", regex=True)
    .str.strip()
)
df
```

```python
# %%time

# Removing NaN values from phone number column
# df["Phone_Number"] = df["Phone_Number"].fillna("")
# df
```

```python
%%time

# Change the phone number column to type string as there are some values that are numbers
df["Phone_Number"] = df["Phone_Number"].astype(str)
df
```

```python
%%time

# Perform regex to remove anything that's not a number from the string
df["Phone_Number"] = (df["Phone_Number"]
.str.replace(r"[^\d]", "", regex=True))
df
```

```python
%%time

# Turn into phone number format with hyphens
df["Phone_Number"] = df["Phone_Number"].str.replace(r"(\d{3})(\d{3})(\d{4})", r"\1-\2-\3", regex=True)
df
```

```python
%%time

# Split Address into 3 columns
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(",", expand=True)
df
```

```python
%%time

# df["Paying Customer"] = df["Paying Customer"].str.replace("Yes", "Y")
# df["Paying Customer"] = df["Paying Customer"].str.replace("No", "N")

df["Paying Customer"] = df["Paying Customer"].replace({"No" : "N", "Yes" : "Y"})
df
```

```python
%%time

# Normalize boolean column
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace({"No" : "N", "Yes" : "Y"})
df
```

```python
# Checking data type of N/A and empty values
df.iloc[18].apply(type)
```

```python
%%time

# Remove all N/A values
df = df.fillna('')
df
```

```python
# Remove the N/a string values since they weren't removed during the .fillna()
df = df.replace("N/a", "")
df
```

### Vectorization
Using vectorized filtering over for looping filtering for faster performance. Especially useful when scaling to larger datasets.

```python
%%time

# Drop row if phone number is blank or they don't want us to contact them
#for x in df.index:
#    if (df.loc[x, "Phone_Number"] == '') or (df.loc[x, "Do_Not_Contact"] in ['Y', '']):
#        df.drop(x, inplace=True)
#df
```

```python
%%time

# Filters and keep all rows where neither condition is true
# then reassigns itself to keep filtered copy
# process through vectorized filtering which is faster
df = df[~((df["Phone_Number"] == "") | (df["Do_Not_Contact"].isin(["Y", ""])))]
df
```

```python
%%time

# Resets the index, making sure to drop the previous indexes
df = df.reset_index(drop=True)
```
